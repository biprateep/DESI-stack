```python
from pathlib import Path

import numpy as np

from bokeh.plotting import figure, show, save
from bokeh.models import Span
from bokeh.io import output_notebook
output_notebook()
```

```python
my_path = Path("/global/cscratch1/sd/bid13/stack_spectra")
file_name = "stacks"+"ELG_1.4_1.5_v0.npz"
```

```python
data = np.load(my_path / file_name)
stack = data["stack"]
wave = data["wave"]
```

```python
flux = stack[0]
```

```python
p = figure(title="Spectra", x_axis_label='Wavelength', y_axis_label='Flux', plot_width=1000, plot_height=300, tooltips=[("Wavelength", "$x"), ("Flux", "$y")]
#            y_range=(-1, 3)
          )
p.line(wave, flux, legend_label="Temp.", line_width=2)

s = Span(location=2795.5, dimension='height')
p.add_layout(s)
s = Span(location=2802.7, dimension='height')
p.add_layout(s)

show(p)
```
